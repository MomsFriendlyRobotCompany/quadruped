```python
def plot_feet(ft):
    def plot_foot(o,i):
        x = [v[0] for v in o]
        y = [v[1] for v in o]
        z = [v[2] for v in o]
        plt.subplot(3,2,i)
        plt.plot(x, label='x');
        plt.plot(y, label='y');
        plt.plot(z, label='z');
        plt.grid(True);
#         plt.axis("equal")
        plt.legend();
    for i, leg in enumerate([1,6,2,5,3,4]):
        plot_foot(ft[leg-1], i+1)
    
plot_feet(ft)
```
